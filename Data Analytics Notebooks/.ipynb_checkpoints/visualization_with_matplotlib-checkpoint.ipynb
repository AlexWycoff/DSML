# Visualization with matplotlib

This is the fun stuff. Matplotlib does SO MUCH. The better you get at using, the better the graphs and visualizations you make will be, and this is a very good thing. Good visualizations are part of good data communication. If people can't follow your graph, they won't follow your point.

That said, beautiful graphs aren't exactly the most academic thing either, so they're not a requirement, just a good life skill.

```python
# Our quintessential libraries right here:
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Again, don't worry about this code. This is not how you'll load in data. 
# I do it this way to keep the DSML github repo clean, you can just sub the path variable for an absolute path
from pathlib import Path
path = Path(globals()['_dh'][0]).parent/"Data/Concussion Injuries 2012-2014.csv"
sports_injuries = pd.read_csv(path)
display(sports_injuries)
```

We can try and make a bar graph for this data, but its not very helpful

```python
teams, counts = np.unique(sports_injuries['Team'], return_counts=True)
plt.bar(teams, counts)
# Remove the overlapping team names
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

A histogram would be a much nicer way to display this

```python
ax = plt.axes()
# labelling the graph
plt.xlabel("Number of injuries")
plt.ylabel("Number of teams")
plt.xlim(4, 22)
# stuff to make the graph look nice
for direction in ['top', 'right', 'left']:
    ax.spines[direction].set_visible(False)
plt.grid(True, axis='y', color='grey')
# actual histogram function
plt.hist(counts, color='purple', zorder=2)
```

Note the intentional use of color. I changed the color from the default blue to purple (for those guessing, it's to be on theme with the Vikings). You can also use gradients (and even make custom ones!) but solid colors work just as well.

Also, do NOT try to memorize all the matplolib functions and variables. Documentation exists for a reason, and if you want to do something but don't know the name, look it up and you'll probably find a stackoverflow post about it (yes, this website is still relevant and won't hallucinate answers).

Looking things up is the best way to learn about matplotlib. I will frequently give advice on how to improve graphs without telling you how to code it (most likely because I don't know either). This is where you can look up the code or borrow it from past projects (save your projects!).

Here's a simpler example with the number of weeks that players are injured for

```python
ax = plt.axes()
plt.xlabel("Number of weeks injured")
plt.ylabel("Number of players")
# Shenanigans
plt.xlim(-0.5, 1.5)
plt.xticks([0, 1])
for direction in ['top', 'right', 'left']:
    ax.spines[direction].set_visible(False)
plt.grid(True, axis='y', color='grey')

# This actually makes the histogram
plt.hist(sports_injuries["Weeks Injured"], color='green', zorder=2, width=0.2)
```
