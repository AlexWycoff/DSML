# Visualization with matplotlib

This is the fun stuff. Matplotlib does SO MUCH. The better you get at using, the better the graphs and visualizations you make will be, and this is a very good thing. Good visualizations are part of good data communication. If people can't follow your graph, they won't follow your point.

That said, beautiful graphs aren't exactly the most academic thing either, so they're not a requirement, just a good life skill.

```python
# Our quintessential libraries right here:
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Again, don't worry about this code. This is not how you'll load in data. 
# I do it this way to keep the DSML github repo clean, you can just sub the path variable for an absolute path
from pathlib import Path
path = Path(globals()['_dh'][0]).parent/"Data/Concussion Injuries 2012-2014.csv"
sports_injuries = pd.read_csv(path)
display(sports_injuries)
```

Now that we have matplotlib, we can make some pretty nice graphs. That histogram of injuries per team from earlier? We can make it into this:

```python
teams, counts = np.unique(sports_injuries['Team'], return_counts=True)
ax = plt.axes()
plt.xlabel("Number of injuries")
plt.ylabel("Number of teams")
plt.xlim(4, 22)
for direction in ['top', 'right', 'left']:
    ax.spines[direction].set_visible(False)
plt.grid(True, axis='y', color='grey')
plt.hist(counts, color='purple', zorder=2)
```

Yeah, that looks way better. Don't be pressured to make graphs that look this good. I copied this graph code from a previous project I spent a lot of time on.

However, do note the intentional use of color. I changed the color from the default blue to purple (for those guessing, it's to be on theme with the Vikings). You can also use gradients (and even make custom ones!) but solid colors work just as well.
