### Searching

Now we can get into the useful stuff: searching and sorting. We use df.loc[index, column] to access an entry in the dataframe. If you want just a row you can use df.loc[index] and df[column] for just a column. Indices are usually integers and columns are usually strings.

Bonus: if for some reason you want to be consistent, you can use df.loc[:, column]. It's the exact same as df[column]. The ":" is just the same as indexing all of the rows, its basically implying 0 : len(df).

```python
# Let's see what teams are represented in this dataset
sports_injuries['Team']
```

Wow, 391 is a lot. However will we sort through this? Numpy comes to the rescue here with np.unique(). It returns all of the unique values in the column.

```python
unique_array = np.unique(sports_injuries['Team'])
print(unique_array)
# This is an f-string. Don't worry about it right now.
print(f"Length of array: {len(unique_array)}")
# Equivalent and less fancy code: print("Length of array:", len(unique_array))
```

That's a lot less! Only 32 unique teams are represented in the dataset. We can also find out how often each is represented by returning counts.

```python
unique_array, team_counts = np.unique(sports_injuries['Team'], return_counts=True)
# If you're wondering how to remember the variables names for function calls like this, 
# they are always in camel_case and never use caps (very unlike Java)

print(unique_array)
# print statements automatically start a new line (also unlike Java)
print(team_counts)
```

This is a pretty interesting distribution. Well, at least it would be if we could make any sense of it. This isn't the topic for today, but we're gonna use matplotlib for this. Luckily, we don't have to import it right now because pandas has some built-in functions to plot histograms for us.

```python
# We have to convert from a numpy array to a Pandas series so we can use Pandas functions (this is important!)
# Pandas series are NOT the same as dataframes. arrays -> series, 2D arrays -> dataframes
team_counts = pd.Series(team_counts)
team_counts.hist()
```

Now we can get a good look. It seems like there's a few teams with unusually high injuries that skew our distribution right, but otherwise its a pretty normal (get it?) distribution.

You'll learn later that this is not a great way to make graphs. This histogram is super ugly (and we use box plots for statistical distributions), but it will do for now.

This dataset has a lot of other interesting features that we'll explore later. Let's get back on track and do some sorting

### Sorting

```python
# Sorting in pandas has some seemingly weird syntax but it will make more sense in a second
# Let's say we're interested in the Minnesota Vikings, we can find all their rows in the dataset like this:
vikings_rows = sports_injuries[sports_injuries['Team'] == 'Minnesota Vikings']
display(vikings_rows)
```

It works as intended, but why did it work that way? sports_injuries['Team'] == 'Minnesota Vikings' actually returns a list with True and False values like this:

```python
display(sports_injuries['Team'] == 'Minnesota Vikings')
```

So we use this list to index the sports_injuries dataframe, giving us only the True rows where the team is the Minnesota Vikings. That's how we get to:

```python
display(sports_injuries[sports_injuries['Team'] == 'Minnesota Vikings'])
```

That's pretty much the toughest thing about sorting with Pandas. Now we can add in some fun conditionals. Let's get all the rows where the Vikings win:

```python
# Take note of the "&" operator (equivalent of Java "&&")
winning_rows = sports_injuries[(sports_injuries['Team'] == 'Minnesota Vikings') & (sports_injuries['Winning Team?'] == 'Yes')]
display(winning_rows)
```

```python
# Another f-string. If you want to use these, put f before the string (hence the name),
# and add your expression in curly braces {} inside the string
print(f"Vikings win percentage (in this dataset): {len(winning_rows) / len(vikings_rows)}")
```

It seems like when we have injured players we don't tend to win. Takes notes if you're interested in sports analytics.

This is the kind of analysis we want to extend to the entire dataset. Maybe we could answer the question "do football teams win less often when a player gets injured in a game?" This isn't the kind of question that just the Vikings can answer, but we could repeat this process for the entire dataset, then compare it to the average win rate of the teams included in the same years (2012-2014). 

This is why Pandas is such a useful tool. It took barely any code to do all of this! Of course, this will not be the case when we have to preprocess our data, but at least the analysis is nice.
